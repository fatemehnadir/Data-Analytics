```python
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
import time
import re
```

```python
NAN = None
class Crawler:
    def __init__(self):
        self.mainDriver = webdriver.Chrome()
        self.data = {'title' :[], 'year':[], 'rating' :[], 'certificate' :[], 'metascore' :[], 'genre' :[], 'time' :[], 'director' :[], 'star1': [], 'star2': [], 'star3': [], 'star4':[]}
    
    def get_title(self, index):
        try:
            element = self.mainDriver.find_element(By.XPATH, f'//*[@id="main"]/div/div[3]/div/div[{index}]/div[3]/h3/a')
            return element.text
        except:
            return NAN
    def get_year(self, index):
        try:
            element = self.mainDriver.find_element(By.XPATH, f'//*[@id="main"]/div/div[3]/div/div[{index}]/div[3]/h3/span[2]')
            year = re.findall(r'\d+', element.text)[0]
            return int(year)
        except:
            return NAN

    def get_rating(self, index):
        try:
            element = self.mainDriver.find_element(By.XPATH, f'//*[@id="main"]/div/div[3]/div/div[{index}]/div[3]/div/div[1]/strong')
            rating = float(element.text)
            return rating
        except:
            return NAN

    def get_certificate(self, index):
        try:
            element = self.mainDriver.find_element(By.XPATH, f'//*[@id="main"]/div/div[3]/div/div[{index}]/div[3]/p[1]/span[1]')
            return element.text
        except:
            return NAN    

    def get_metascore(self, index):
        try:
            element = self.mainDriver.find_element(By.XPATH, f'//*[@id="main"]/div/div[3]/div/div[{index}]/div[3]/div/div[3]/span')
            return int(element.text)
        except:
            return NAN    
           
    def get_genre(self, index):
        try:
            element = self.mainDriver.find_element(By.XPATH, f'//*[@id="main"]/div/div[3]/div/div[{index}]/div[3]/p[1]/span[5]')
            genres = element.text.split(',')
            return genres[0]
        except:
            return NAN    
           
    def get_time(self, index):
        try:
            element = self.mainDriver.find_element(By.XPATH, f'//*[@id="main"]/div/div[3]/div/div[{index}]/div[3]/p[1]/span[3]')
            time = re.findall(r'\d+', element.text)[0]
            return int(time)
        except:
            return NAN    
           
    def get_director(self, index):
        try:
            element = self.mainDriver.find_element(By.XPATH, f'//*[@id="main"]/div/div[3]/div/div[{index}]/div[3]/p[3]/a[1]')
            return element.text
        except:
            return NAN    

    def get_star(self, index, number):
        try:
            element = self.mainDriver.find_element(By.XPATH, f'//*[@id="main"]/div/div[3]/div/div[{index}]/div[3]/p[3]/a[{number}]')
            return element.text
        except:
            return NAN    


          
    def run(self):
        for i in range(20):
            self.mainDriver.get(f"https://www.imdb.com/search/title/?groups=top_1000&start={50*i+1}&ref_=adv_nxt")
            for j in range(1, 51):
                self.data['title'].append(self.get_title(j))
                self.data['year'].append(self.get_year(j))
                self.data['rating'].append(self.get_rating(j))
                self.data['certificate'].append(self.get_certificate(j))
                self.data['metascore'].append(self.get_metascore(j))
                self.data['genre'].append(self.get_genre(j))
                self.data['time'].append(self.get_time(j))
                self.data['director'].append(self.get_director(j))
                for k in range(1, 5):
                    self.data[f'star{k}'].append(self.get_star(j, k))
        self.mainDriver.close()    
        return pd.DataFrame(self.data)
crawler = Crawler()
data = crawler.run()

data.to_csv('IMDB.csv', index=False, index_label=False)
```
